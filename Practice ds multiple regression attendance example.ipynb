```python
import numpy as np #import for numerical operations , just in case
import pandas as pd #import for numerical operations, neccessary library 
import matplotlib.pyplot as plt #import for plottinng lines and points from csv files
import statsmodels.api as sm #import to add constant to the regression equation b0
import seaborn as sns #import for beautifying the graph
sns.set() #command to convert all statsmodels commands to seaborn commands without having to know seaborn syntax
```

```python
raw_data = pd.read_csv('/Users/aryamonani/Desktop/Udemy data sci course/The Data Science Course 2018 - All Resources/Part_5_Advanced_Statistical_Methods_(Machine_Learning)/S33_L204/1.03. Dummies.csv')
```

```python
raw_data #shows the data in a tabular form , next step will be to convert all yes and no to 0s and 1s so that they can be used as a variable for computation
```

```python
data = raw_data.copy() #copy raw_data's content into data
```

```python
data #data shown in a tabular form
```

```python
data['Attendance'] = data['Attendance'].map({'No':0,'Yes':1}) #mapping yes and no in the attendance section to 0s and 1s
```

```python
data #showing the table in 0s and 1s
```

```python
data.describe() #describing data in a more detailed format
```

```python
y = data['GPA'] #defining primary variables in the linear regression equation
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(data['SAT'],y)
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
yhat = 0.0017*data['SAT'] + 0.275
fig = plt.plot(data['SAT'],yhat_no,lw=2,c='orange')
fig = plt.plot(data['SAT'],yhat_yes,lw=2,c='red')
fig = plt.plot(data['SAT'],yhat,lw=2,c='blue')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```
